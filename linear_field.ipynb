```python
import numpy as np
```

```python
a = [
    [-4, 1, 0, 1, 0, 0, 0, 0],
    [1, -4, 1, 0, 0, 0, 0, 0],
    [0, 1, -4, 0, 1, 0, 0, 0],
    [1, 0, 0, -4, 0, 1, 0, 0],
    [0, 0, 1, 0, -4, 0, 0, 1],
    [0, 0, 0, 1, 0, -4, 1, 0],
    [0, 0, 0, 0, 0, 1, -4, 1],
    [0, 0, 0, 0, 1, 0, 1, -4],
    [0, 1, 0, 1, 1, 0, 1, 0]
    ]

b = [0, 1, 0, 1, 1, 0, 1, 0, -1]

a = np.array(a)
b = 4*np.array(b)
```

```python
a_inv = np.linalg.inv(a)
```

```python
a_inv
```

```python
x = np.matmul(a_inv, -np.vstack(b))
x
```

```python
(np.matmul(a, x) + b)
```

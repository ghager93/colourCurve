```python
import numpy as np
import matplotlib.pyplot as plt

from drawing import *

import importlib
```

```python
def draw_curve(curve, n_points=20):
    plt.scatter(*[p for p in zip(*curve.get_equidistant_points(n_points))])
```

```python
line = ParameterisedLine([0, 1], [2, 4])
```

```python
points = line.get_equidistant_points(20)
```

```python
plt.scatter(*[p for p in zip(*points)])
```

```python
circle = ParameterisedCircle(4, [0, 0])
points = circle.get_equidistant_points(20)
plt.scatter(*[p for p in zip(*points)])
```

```python
curve = ParameterisedEllipse(4, 10, [0, 0])
draw_curve(curve)
```

```python
curve = ParameterisedRotatedEllipse(4, 10, [0, 0], 45 * np.pi / 180)
draw_curve(curve)
```

```python
curve = ParameterisedBezier([[0, 0], [1, 1], [3, -2], [2, 2]])
draw_curve(curve)
plt.scatter([0, 1, 3, 2], [0, 1, -2, 2])
```

```python
curve = ParameterisedRationalBezier([[0, 0], [1, 1], [0, 2]], [1, -0.1, 1])
draw_curve(curve)
plt.scatter([0, 1, 0], [0, 1, 2])
```

```python
curve = ParameterisedPiecewiseLine([[0, 0], [1, 0], [0, 1]])
draw_curve(curve)
```

```python
curve = ParameterisedBezier([[0, 0], [1, 1], [0, 1], [2, 0], [3, -3], [4, 4], [5, 5], [0, 0]])
draw_curve(curve, 50)
```

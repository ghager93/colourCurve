```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
%matplotlib notebook
```

```python
def _gaussian3d_func(t, x0):
    t = np.atleast_1d(t)[:, None, None]
    x0 = np.atleast_1d(x0)[:, None, None]
    
    return lambda x: np.exp(-np.sum((x - x0)**2 / (2 * t**2), axis=0))
```

```python
def _weight_func(t, px, py):
    t, px, py = np.atleast_1d(t, px, py)
    
    def _w(x):
        dx = [di(x) for di in d]
        out = dx
        for i in range(1, len(d)):
            out *= 1 - np.roll(dx, i, axis=0)
            
        return out
    
    d = [_gaussian3d_func(t, np.array([px[i], py[i]])) for i in range(len(px))]
    
    return _w
```

```python
def _sample_vector_normals(px, py, qx, qy):
    px, py, qx, qy = np.atleast_1d(px, py, qx, qy)
    
    out = np.array([qx - px, qy - py])
    return out / np.linalg.norm(out, axis=0)
```

```python
def _velocity_func(t, px, py, qx, qy):
    def _v(x):
        wx = _weight_func(t, px, py)(x)
        return np.sum(wx[None, :, :, :] * v_norms[:, :, None, None], axis=1) / np.sum(wx, axis=0)
        
    v_norms = _sample_vector_normals(px, py, qx, qy)
    
    return _v
```

```python
px = np.array([0.2, 0.45, 0.7])
py = np.array([0.4, 0.3, 0.3])
t = [1, 1]

qx = np.array([0.4, 0.3, 0.7])
qy = np.array([0.2, 0.7, 0.1])

x1 = np.linspace(0, 1, 11)
y1 = np.linspace(0, 1, 11)
x = np.meshgrid(x1, y1)
```

```python
w = _weight_func(t, px, py)
```

```python
wx = w(x)
wx.shape
```

```python
vdx, vdy = [qx - px, qy - py] / np.linalg.norm([qx - px, qy - py], axis=0)
vd = np.array([vdx, vdy])
```

```python
vdx, vdy = wx[None, :, : ,:] * vd[:, :, None, None]
vd = np.array([vdx, vdy])
```

```python
plt.imshow(np.sum(wx, axis=0))
plt.figure()
plt.imshow(wx[0])
```

```python
s = x + np.sum(vd, axis=1) / np.sum(wx, axis=0)
```

```python
plt.quiver(*x, *(s-x), color='blue', scale_units='xy', scale=10)
plt.quiver(px, py, qx-px, qy-py, color='red', scale_units='xy', scale=1)
```

```python
px, py = [0.3, 0.6], [0.3, 0.6]
qx, qy = [0.6, 0.3], [0.6, 0.3]

px, py, qx, qy = np.atleast_1d(px, py, qx, qy)

v = _velocity_func(t, px, py, qx, qy)
```

```python
plt.quiver(*x, *v(x), color='blue', scale_units='xy', scale=10)
plt.quiver(px, py, qx-px, qy-py, color='red', scale_units='xy', scale=1)
```

```python
px, py = [0.1, 0.5, 0.4], [0.3, 0.6, 0.7]
qx, qy = [0.6, 0.5, 0.2], [0.8, 0.3, 0.3]

px, py, qx, qy = np.atleast_1d(px, py, qx, qy)
pq = np.linalg.norm([qx - px, qy - py], axis=0)

v = _velocity_func(t, px, py, qx, qy)
```

```python
gauss_list = np.array([_gaussian3d_func(0.3, [px[i], py[i]]) for i in range(len(px))])
```

```python
gx = np.max([pq[i] * gauss_list[i](x) for i in range(len(pq))], axis=0)
gx.shape
```

```python
x[0].shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(px, py, pq, marker='o', color='black', linewidths=20)

vx = np.linalg.norm(v(x), axis=0)

# Normalize the colors based on Z value
norm = plt.Normalize(vx.min(), vx.max())
colors = cm.jet(norm(vx))
surf = ax.plot_surface(x[0], x[1], vx, facecolors=colors, shade=False)
surf.set_facecolor((0,0,0,0))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(px, py, pq, marker='o', color='black', linewidths=20)

# Normalize the colors based on Z value
norm = plt.Normalize(gx.min(), gx.max())
colors = cm.jet(norm(gx))
surf = ax.plot_surface(x[0], x[1], gx, facecolors=colors, shade=False)
surf.set_facecolor((0,0,0,0))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(px, py, pq, marker='o', color='black', linewidths=20)

vx = np.linalg.norm(gx * v(x), axis=0)

# Normalize the colors based on Z value
norm = plt.Normalize(vx.min(), vx.max())
colors = cm.jet(norm(vx))
surf = ax.plot_surface(x[0], x[1], vx, facecolors=colors, shade=False)
surf.set_facecolor((0,0,0,0))
```
